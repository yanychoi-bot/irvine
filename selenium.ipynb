```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# get selenium driver and dependencies
```

```python
class Job:

    title: str
    company: str
    location: str
    salary: str
    job_type: str

    def __init__(self, title, company, location, salary, job_type):
        self.title = title
        self.company = company
        self.location = location
        self.salary = salary
        self.job_type = job_type
    
    def __str__(self):
        return f"{self.title} | {self.company} | {self.location} | {self.salary}"
```

```python
dr = webdriver.Chrome() # use chrome webdriver for selenium
```

```python
from time import sleep

# get indeed job search page with multiple options

INDEED_BASE_URL = 'https://www.indeed.com/jobs'

LOCATION = 'California'

LIMIT = 3000


    


jobs: [Job] = []


temp = webdriver.Chrome()


def xpath(element):
    try:
        return temp.find_element(By.XPATH, element).text
    except:
        return ""

for i in range(0, LIMIT, 10):
    OFFSET = i
    try:
        dr.get(f'{INDEED_BASE_URL}?l={LOCATION}&start={OFFSET}')

        urls = dr.find_elements(By.CLASS_NAME, "jcs-JobTitle")

        for url in urls:
            print(url.get_attribute('href'))
            try:
                temp.get(url.get_attribute('href'))
                sleep(1)
                title = xpath("//h1[@data-testid='jobsearch-JobInfoHeader-title']")
                company = xpath("//div[@data-testid='inlineHeader-companyName']")
                location = xpath("//div[@data-testid='inlineHeader-companyLocation']")
                salary = xpath("//*[@id='jobDetailsSection']/div[2]/div[2]/span/div/div/div")
                job_type = xpath("//*[@id='jobDetailsSection']/div[3]/div[2]/div/div/div/div")
                jobs.append(Job(title, company, location, salary, job_type))
                print(title)
            except:
                pass
    except:
        pass

```

```python
import re

def get_hourly_rate(rate, time_unit):
    if time_unit == 'hour':
        return rate
    if time_unit == 'month':
        return rate / 160
    if time_unit == 'year':
        return rate / 1920

def get_salary(text):
    pattern = r'\$(\d.+) - \$(\d.+) (an|a) (hour|month|year)|\$(\d.+) (an|a) (hour|month|year)'

    match = re.search(pattern, text)

    if match:
        if match.group(1) is not None:
            lower_range = float(match.group(1).replace(',', ''))
            upper_range = float(match.group(2).replace(',', ''))

            hourly_rate = ((upper_range + lower_range) / 2)
            time_unit = match.group(4)
            return get_hourly_rate(hourly_rate, time_unit).__round__(2)
        if match.group(5) is not None:
            hourly_rate = match.group(5)
            time_unit = match.group(7)
            return get_hourly_rate(hourly_rate, time_unit)
```

```python
import csv
print(len(jobs))

with open('jobs.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['Title', 'Company', 'Location', 'Salary'])
    for job in jobs:
        try:
            writer.writerow([job.title, job.company, job.location, get_salary(job.salary)])
        except:
            pass
```
